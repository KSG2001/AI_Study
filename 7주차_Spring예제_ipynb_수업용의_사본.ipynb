```python
import numpy as np
```

```python
force = np.array([1.1, 1.9, 3.2, 4.4, 5.9, 7.4, 9.2])
target = np.array([1.5, 2.1, 2.5, 3.3, 4.1, 4.6, 5.0])
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(force, target, random_state = 47)
```

```python
from sklearn.linear_model import LinearRegression
lreg = LinearRegression()

train_input = train_input.reshape(-1,1)
test_input =  test_input.reshape(-1,1)

lreg.fit(train_input, train_target)
```

```python
print(lreg.coef_, lreg.intercept_)
print(lreg.predict([[5.0]]))
```

```python
import matplotlib.pyplot as plt

plt.scatter(force, target)
plt.scatter([5.0], [3.43], marker='^')

plt.show()
```

```python
lreg.score(test_input, test_target)
```

```python
plt.scatter(train_input, train_target)
plt.scatter(test_input, test_target, marker='D')
plt.plot([1.1, 9.2], [1.1*lreg.coef_+lreg.intercept_, 9.2*lreg.coef_+lreg.intercept_])

plt.show()
```

```python
print(train_input)
print(train_target)
```

```python
force_poly = np.column_stack((train_input**2, train_input))
test_poly = np.column_stack((test_input**2, test_input))

print(force_poly)
print(test_input)
```

```python
lregPoly = LinearRegression()
lregPoly.fit(force_poly,train_target)

print(lregPoly.coef_, lregPoly.intercept_)
```

```python
point=np.arange(0,12)

plt.scatter(train_input, train_target)
plt.scatter(test_input, test_target)
plt.plot(point, -0.029*point**2+0.735*point+0.709)

plt.show()
```

```python
force_poly3 = np.column_stack((train_input**3, train_input**2, train_input))
test_poly3 = np.column_stack((test_input**3, test_input**2, test_input))

lregPoly3 = LinearRegression()
lregPoly3.fit(force_poly3,train_target)

print(lregPoly3.coef_, lregPoly3.intercept_)
```
