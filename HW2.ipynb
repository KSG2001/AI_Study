```python
import cv2
import matplotlib.pyplot as plt
```

```python
if __name__=='__main__':
    img = cv2.imread(r"/content/gdrive/My Drive/computervision/lucky.jpg")
    avg_img = cv2.boxFilter(img, -1, (3,3), normalize=True)


    plt.subplot(1, 2, 1)
    plt.title("original")
    plt.imshow(img)

    plt.subplot(1, 2, 2)
    plt.title("avg")
    plt.imshow(avg_img)

    plt.show()
```

#수평에지 opencv의 **filter2D( )**함수 사용

```python
import numpy as np
```

```python
if __name__ == '__main__':
    img_path = 'r/content/gdrive/My Drive/computervision/lucky.jpg'
    img_2D = cv2.imread(img_path)

    if img_2D is not None :
        kernel = np.ones((5, 5), dtype=np.float64)/25.
        kernel[0, :] = -2
        kernel[1, :] = -1
        kernel[2, :] = 0
        kernel[3, :] = 1
        kernel[4, :] = 2
        print(kernel)

        filter2D_img = cv2.filter2D(img_2D, -1, kernel)

        cv2.imshow('src', filter2D_img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
```

#opencv의 **GaussianBlur( )**함수 사용, sigma=1, 3, 5 세 경우 모두 화면 표시


```python
img_path = 'r/content/gdrive/My Drive/computervision/lucky.jpg'
img = cv2.imread(img_path,  cv2.COLOR_BGR2RGB)

if img is not None:
    gaussian_img1 = cv2.GaussianBlur(img, (0, 0), 1)
    gaussian_img2 = cv2.GaussianBlur(img, (0, 0), 2)
    gaussian_img3 = cv2.GaussianBlur(img, (0, 0), 3)

    plt.subplot(2, 2, 1)
    plt.title("original")
    plt.imshow(img)

    plt.subplot(2, 2, 2)
    plt.title("sigma = 1")
    plt.imshow(gaussian_img1)

    plt.subplot(2, 2, 3)
    plt.title("sigma = 2")
    plt.imshow(gaussian_img2)

    plt.subplot(2, 2, 4)
    plt.title("sigma = 3")
    plt.imshow(gaussian_img3)

    plt.show()
```
