```python
import pandas as pd
import matplotlib.pyplot as plt

emotion = pd.DataFrame([
        ['angry', 976, 0, 0, 0, 6, 18, 0],
        ['disgusted', 0, 997, 0, 0, 3, 0, 0],
        ['fearful', 1, 0, 982, 0, 0, 6, 11],
        ['happy', 1, 2, 2, 995, 0, 0, 0],
        ['neutral', 14, 0, 0, 0, 975, 11, 0],
        ['sad', 17, 0, 0, 0, 5, 978, 0],
        ['surprised', 0, 0, 3, 0, 0, 0, 997],
    ], columns=['actual/predicted', 'angry', 'disgusted', 'fearful', 'happy', 'neutral', 'sad', 'surprised'])   # 데이터


fig, ax = plt.subplots()                                # heat map만들기, subplots() 을 사용하여 fig(ure; 프레임), ax(is; 축)을 만들자.
map = ax.imshow(emotion.iloc[:, 1:], cmap='binary')      # 

ax.set_xticks(range(len(emotion.columns[1:])))          # x, y axis 설졍
ax.set_xticklabels(emotion.columns[1:])
ax.set_yticks(range(len(emotion)))
ax.set_yticklabels(emotion.iloc[:, 0])

ax.set_xlabel('Predicted Labels')                       #라벨 넣기
ax.set_ylabel('True Labels')


for i in range(len(emotion)):                                                                             # 셀 값 입력
    for j in range(1, len(emotion.columns)):
        text_color = 'white' if emotion.iloc[i, j] > 900 else 'black'
        ax.text(j-1, i, emotion.iloc[i, j], ha="center", va="center", color=text_color, fontsize=10)

plt.show()    # show
```
